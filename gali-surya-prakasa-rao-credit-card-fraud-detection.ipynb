```python
#as we obserbed in the above plot the distribution is not a normal distribution so we have perform a normalization. but before that lets seperate the class column with other colums as class column is our outpu column
```

```python
x = df.drop(columns=['Class'],axis=1)
y = df['Class']
```

```python
#for doing the normalization we can use log transformtion or else the standard scalar method in the sklearn preprocessing module. as we do not have the chategerical columns we can eliminate the encoding of the data 
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()


z =ss.fit_transform(x)
    
    
```

```python
#now lets do the train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(z,y,test_size = 0.25)
```

```python
from sklearn.metrics import classification_report,f1_score
def classify(model,x_train,x_test,y_train,y_test):
    model.fit(x_train,y_train)
    result =model.predict(x_test)
    report = classification_report(result,y_test)
    print(report)
    print(f1_score(result,y_test))
```

```python
from sklearn.linear_model import LogisticRegression
model= LogisticRegression()
classify(model,x_train,x_test,y_train,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
#from sklearn.model_selection import GridSearchCV
#parameters ={'n_estimators':[400,500,600],'max_depth':[2,3,4]}
model1 = RandomForestClassifier(n_estimators=400,max_depth = 4)
classify(model1,x_train,x_test,y_train,y_test)
```
