```python
plt.style.use('fivethirtyeight')
fig, ax = plt.subplots(ncols = 2, nrows =4, figsize = (20,20))
ax = ax.flatten()
index =0
for col,values in df.items():
  sns.countplot(y = col, data = df,ax = ax[index])
  plt.tight_layout(pad = 0.5,w_pad = 0.6,h_pad = 5.0)
  index+=1
  plt.xticks(rotation =90)
```

```python
plt.style.use('fivethirtyeight')
fig,ax = plt.subplots(ncols =2,nrows = 4,figsize =(20,20))
ax= ax.flatten()
index= 0
for col,values in df.items():
  sns.distplot(values,ax =ax[index])
  plt.tight_layout(pad = 0.5,w_pad = 0.7,h_pad = 5.0)
  index+=1
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
df1 =ss.fit_transform(df)
```

```python
from sklearn.cluster import KMeans
d = []
for i in range(6):
  model = KMeans(n_clusters = i,init = "k-means++")
  y = model.fit(df1)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
df2 = pd.DataFrame(pca.fit_transform(df1))
```

```python
df2
```

```python
import scipy.cluster.hierarchy as hier
dendrogram = hier.dendrogram(hier.linkage(df2,method ='ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering
model = AgglomerativeClustering(n_clusters=2,affinity = 'euclidean',linkage= 'ward')
y =model.fit_predict(df2)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components =3)
df3 = pd.DataFrame(pca.fit_transform(df1))
```

```python
import scipy.cluster.hierarchy as hier
dendrogram = hier.dendrogram(hier.linkage(df3,method = 'ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering
model = AgglomerativeClustering(n_clusters = 3,affinity = 'euclidean',linkage = 'ward')
y = model.fit_predict(df3)
df3['predict'] = y 
```

```python
df3
```

```python
# scatterplot
plt.style.use('fivethirtyeight')
fig =plt.figure(figsize = (10,10))
ax = fig.add_subplot(111,projection = '3d')
ax.scatter(df3[0][df3['predict']==0],df3[1][df3['predict']==0],df3[2][df3['predict']==0],c = 'red')
ax.scatter(df3[0][df3['predict']==1],df3[1][df3['predict']==1],df3[2][df3['predict']==1],c = 'blue')
ax.scatter(df3[0][df3['predict']==2],df3[1][df3['predict']==2],df3[2][df3['predict']==2],c = 'green')

```
