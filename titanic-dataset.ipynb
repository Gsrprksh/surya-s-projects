```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(train_1,train_2,test_size= 0.25)
```

```python
from sklearn.model_selection import cross_val_score
def classify(model,x_train,x_test,y_train,y_test,train_1,train_2):
    model.fit(x_train,y_train)
    model.score(x_test,y_test)
    print(model.score(x_test,y_test))
    p = cross_val_score(model,train_1,train_2)
    print(np.mean(p)*100)
   
```

```python
np.all(np.isfinite(x_test['Fare']))
x_test.replace([np.inf,-np.inf],np.nan,inplace= True)
#x_train.replace([np.inf, -np.inf], np.nan, inplace=True)
#x_train.isnull().sum()
#x_train['Fare']= x_train['Fare'].fillna(x_train['Fare'].mean())
#x_train.isnull().sum()
x_test['Fare']= x_test['Fare'].fillna(x_test['Fare'].mean())
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train,y_train)
print(model.score(x_test,y_test))
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
classify(model,x_train,x_test,y_train,y_test,train_1,train_2)
```

```python
from sklearn.ensemble import RandomForestClassifier
model1 = RandomForestClassifier()
classify(model1,x_train,x_test,y_train,y_test,train_1,train_2)
```

```python
parameters ={'n_estimators':[400,500,600],'max_depth' :[2,3,5]}
from sklearn.metrics import mean_absolute_error
model1.fit(x_train,y_train)
pred = model1.predict(x_test)
mean_absolute_error(pred,y_test)
```
